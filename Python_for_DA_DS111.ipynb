import random as r 

opt = ["kéo", "búa", "bao"]

print("Chọn 1 trong 3 lựa chọn: kéo, búa, bao")

player = input("Lựa chọn của bạn: ").strip().lower() #.strip() để bỏ khoảng trắng 2 đầu, .lower() để chuyển về chữ thường

if player not in opt:
    print("Lựa chọn không hợp lệ!")
else:
    machine = r.choice(opt)
    print("Lựa chọn của máy: ", machine)
    
    if player == machine:
        print("Kết quả hòa!")
    elif(player == "kéo" and machine == "bao") or \
        (player == "búa" and machine == "kéo") or \
        (player == "bao" and machine == "búa"):
        print("Bạn đã thắng!")
    else:
        print("Bạn đã thua!")

### **2. List Comprehension**

• Tạo một danh sách bằng cách kết hợp lệnh for và biểu thức

<u>**Cú pháp – còn gọi là cú pháp comprehension**</u>

<p align="center">new_list = [expression for item in iterable if condition]</p>

- expression: biểu thức để tính toán giá trị của từng phần tử trong list mới.
- item: biến được sử dụng để lặp qua từng phần tử trong list ban đầu.
- condition: điều kiện kiểm tra mà các item phải thỏa để được thêm vào new_list
ds = [5, -3, 6, -2, 8, -1, 4, -7]
new_ds = [item for item in ds if item > 0]
print(new_ds)
