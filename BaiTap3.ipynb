# Số lượng người nhập cư từ toàn thế giới đến Canada theo từng năm
# Tính tổng tất cả các quốc gia cho mỗi năm (cột)

```python
migrating_data = df.loc[:, 1980:2013].sum()
plt.title('Số lượng người nhập cư từ toàn thế giới đến Canada (1980-2013)')
plt.ylabel('Số lượng người nhập cư')
plt.xlabel('Năm')
plt.plot(migrating_data.index, migrating_data.values)
```

```python
sns.lineplot(data=df.loc[:, 1980:2013].sum())
```

# Chọn ra 5 quốc gia có số lượng di cư nhiều nhất Area plot

```python
total_by_country = df.loc[:, 1980:2013].sum(axis=1)

# Lấy top 5 quốc gia có số lượng nhập cư nhiều nhất
top5_countries = total_by_country.nlargest(5)
print("Top 5 quốc gia có số lượng nhập cư nhiều nhất:")
top5_countries_index = top5_countries.index
```

```python
df.loc[top5_countries_index, 1980:2013].T.plot(kind='area', figsize=(10, 6), stacked=False)
```

# Histogram - frequency distribution

```python
df.loc[top5_countries_index, 1980:2013].T.plot(kind='area', figsize=(10, 6))
```

```python
canada = df.loc[:, 2010]
```

```python
count, bin_edges = np.histogram(canada)
count
```

```python
bin_edges
```

```python
canada.plot(kind='hist', figsize=(10, 6), bins=50, xticks=bin_edges)
plt.title('Biểu đồ tần số nhập cư từ các quốc gia đến Canada trong năm 2010')
plt.xlabel('Số lượng người nhập cư')
plt.ylabel('Số quốc gia')
plt.show()
```

```python
sns.histplot(canada, bins=50, kde=True)
```
